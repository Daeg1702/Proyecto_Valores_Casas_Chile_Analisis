```python
import pandas as pd
from unidecode import unidecode
import numpy as np
```

```python
df_precio_casas=pd.read_csv(r"data\2023-03-08 Precios Casas RM.csv",sep=",")
comunas = pd.read_csv("data\Latitud - Longitud Chile.csv",sep=",") 
```

```python
comunas.info()
```

```python
df_precio_casas.info()
```

```python
comunas = comunas[comunas["Región"] == "Metropolitana de Santiago"] #Seleccionamos las comunas que están en santiago
comunas.Comuna.unique() #Fíjense en cómo están escritas, con espacios y tildes. 
```

```python
comunas["Comuna"] = comunas.Comuna.str.replace(" ", "") #Vamos a quitarles los espacios
comunas["Comuna"] = comunas["Comuna"].apply(unidecode) #Quitamos caracteres extraños como tildes
comunas["Comuna"] = comunas["Comuna"].str.lower() #Dejamos en minusculas
comunas.head()
```

```python
df_precio_casas["Comuna"] = df_precio_casas["Comuna"].apply(unidecode) #Quitamos caracteres extraños
df_precio_casas["Comuna"] = df_precio_casas["Comuna"].str.lower() #Dejamos en minusculas
df_precio_casas
```

```python
df_precio_casas = df_precio_casas.merge(comunas, on="Comuna", how="left")
df_precio_casas
```

```python
df_precio_casas.columns
```

```python
df_precio_casas = df_precio_casas.drop(columns=["Price_CLP", "Price_UF", "Total Area", "Realtor", "CUT (Código Único Territorial)",
                          "IDH 2005", "Población Año 2017", "Densidad (hab./km2)", "Latitud (Sexagesimal)", "Longitud (Sexagesimal)"," Superficie (km2) ","Latitud (Decimal)",
                          "Longitud (decimal)"])
```

```python

df_precio_casas[[ 'Dorms', 'Baths', 'Parking']] = df_precio_casas[[ 'Dorms', 'Baths', 'Parking']].fillna(0)
```

```python
df_precio_casas[[ 'Dorms', 'Baths', 'Parking']]=df_precio_casas[[ 'Dorms', 'Baths', 'Parking']].astype('Int64')
```

```python
df_precio_casas.info()
```

Rangos de cantidades de Baños

```python
condiciones_bath = [
    df_precio_casas['Baths'].between(1, 5),
    df_precio_casas['Baths'].between(6, 10),
    df_precio_casas['Baths'].between(11, 15),
    df_precio_casas['Baths'].between(16, 20),
    df_precio_casas['Baths'].between(21, 25),
    df_precio_casas['Baths'] > 25
]

seleccion_bath = [
    '1-5',
    '6-10',
    '11-15',
    '16-20',
    '21-25',
    '> 25'
]

df_precio_casas['rango_baths'] = np.select(condiciones_bath, seleccion_bath, default='Sin Baños')
```

Rangos de cantidades de Dormitorios

```python
condiciones_dorm = [
    df_precio_casas['Dorms'].between(1, 5),
    df_precio_casas['Dorms'].between(6, 10),
    df_precio_casas['Dorms'].between(11, 15),
    df_precio_casas['Dorms'].between(16, 20),
    df_precio_casas['Dorms'].between(21, 25),
    df_precio_casas['Dorms'] > 25
]

seleccion_dorm = [
    '1-5',
    '6-10',
    '11-15',
    '16-20',
    '21-25',
    '> 25'
]

df_precio_casas['rango_dormitorios'] = np.select(condiciones_dorm, seleccion_dorm, default='Sin Dormitorios')
```

Rangos de cantidades de Parqueos

```python
condiciones_park = [
    df_precio_casas['Parking'].between(1, 5),
    df_precio_casas['Parking'].between(6, 10),
    df_precio_casas['Parking'].between(11, 15),
    df_precio_casas['Parking'].between(16, 20),
    df_precio_casas['Parking'].between(21, 25),
    df_precio_casas['Parking'] > 25
]

seleccion_park = [
    '1-5',
    '6-10',
    '11-15',
    '16-20',
    '21-25',
    '> 25'
]

df_precio_casas['rango_parqueos'] = np.select(condiciones_dorm, seleccion_dorm, default='Sin Parqueos')
```

```python
df_precio_casas.to_csv("data_carga\precio_casas_chile_bi.csv",index=False)
```
